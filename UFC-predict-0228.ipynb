```python
import pandas as pd

df = pd.read_csv("D:\\Downloads\\UFC.csv")

# Data Cleaning
df['date'] = pd.to_datetime(df['date'])
df['r_dob'] = pd.to_datetime(df['r_dob'])
df['b_dob'] = pd.to_datetime(df['b_dob'])
df['age_diff'] = (df['r_dob']-df['b_dob']).dt.days
df['reach_diff'] = df['r_reach']-df['b_reach']
df['is_winner'] = (df['r_name'] == df['winner']).astype(int)

#grappling features - include takedown accuracy, sub average, takedown defense
df.head(10)
```

```python

import numpy as np
import xgboost as xgb
from sklearn.metrics import accuracy_score, log_loss
from sklearn.preprocessing import StandardScaler


# Sort Chronologically 
df = df.sort_values(by='date').reset_index(drop=True)

# Features included so far
features = [
    'r_splm', 'r_sapm', 'r_td_avg_acc','r_sub_avg','r_td_def',
    'b_splm', 'b_sapm', 'b_td_avg_acc',  'b_sub_avg',  'b_td_def', 
    'reach_diff', 'age_diff'
]
target = 'is_winner' # Binary Variable

# Walk-Forward Validation Setup
start_index = 2000 
batch_size = 500 # We will predict the next 500 fights, then retrain.

predictions = []
actuals = []
probabilities = []

print(f"Starting Walk-Forward Validation on {len(df)} fights...")

for i in range(start_index, len(df), batch_size):
    # Iteration Step - This splits up the data for the Walk-Forward Validation 
    train_data = df.iloc[:i]
    test_data = df.iloc[i : i + batch_size]
    
    if test_data.empty:
        break
        
    X_train = train_data[features]
    y_train = train_data[target]
    
    X_test = test_data[features]
    y_test = test_data[target]
    
    # Model Training
    model = xgb.XGBClassifier(
        #try tuning some of these guys
        #Model Comparison: While raw values are hard to interpret in isolation, lower log loss is always better when comparing models.
        #Calibration: A low log loss indicates the model's predicted probabilities align with the actual frequency of outcomes.
        n_estimators=100, 
        max_depth=3, 
        learning_rate=0.1, 
        objective='binary:logistic',
        eval_metric='logloss',
        use_label_encoder=False
    )
    #give random forest a chance
    #feature randomness? 
    #what is regularization
    model.fit(X_train, y_train)
    
    # Predictions on test set
    batch_probs = model.predict_proba(X_test)[:, 1]
    batch_preds = (batch_probs > 0.5).astype(int) # Threshold at 50% - since we predict the probability of the red fighter winning, 50% makes perfect sense for classification. 
    
    # Storing Results
    probabilities.extend(batch_probs)
    predictions.extend(batch_preds)
    actuals.extend(y_test.values)
    
    print(f"Batch {i//batch_size}: Trained on {len(train_data)} fights, predicted {len(test_data)}")

# Final Evaluation
overall_acc = accuracy_score(actuals, predictions)
overall_logloss = log_loss(actuals, probabilities)

print("\n--- Final Model Performance ---")
print(f"Accuracy: {overall_acc:.2%}") 
print(f"Log Loss: {overall_logloss:.4f}")
#log loss of roughly 0.693 - coin flip

# Feature Importance
import matplotlib.pyplot as plt

xgb.plot_importance(model, importance_type='weight')
plt.title("What Drives the Fight Outcome?")
plt.show()
```
