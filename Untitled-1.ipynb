```python
import pandas as pd

df = pd.read_csv("D:\\Downloads\\UFC.csv")
#dependent variable - red corner victory? (binary)
#what are the most important metrics for predicting fight outcomes
#style matchups, strike accuracy vs strike absorption, height, fight finishing methods, knockdown to significant strike ratio ('punch power'), sub attempts per minute
#ground strikes per minute, takedown defence/accuracy, weight class, win to decision ratio ('decision merchant' - binary?), time since last fight
#outcome of last fight (did this mf get KOd and come back too soon?)

df['date'] = pd.to_datetime(df['date'])
df['r_dob'] = pd.to_datetime(df['r_dob'])
df['b_dob'] = pd.to_datetime(df['b_dob'])
df['age_diff'] = (df['r_dob']-df['b_dob']).dt.days
df['reach_diff'] = df['r_reach']-df['b_reach']
df['is_winner'] = (df['r_name'] == df['winner']).astype(int)

df.head(10)
```

```python

import numpy as np
import xgboost as xgb
from sklearn.metrics import accuracy_score, log_loss
from sklearn.preprocessing import StandardScaler

# 1. Load the "Model Ready" Data - needs cleaning


# 2. Sort Chronologically (CRITICAL for Walk-Forward)
df = df.sort_values(by='date').reset_index(drop=True)

# 3. Define Features & Target
# define strikes landed vs opponent strikes absorbed and vice versa 
# We exclude 'odds' from features because we use them for betting logic, not prediction.
features = [
    'r_splm', 'r_sapm', 
    'b_splm', 'b_sapm',
    'reach_diff', 'age_diff' # Assuming these exist from your earlier steps
]
target = 'is_winner' # Assumes encoded: 1 = Red, 0 = Blue

# 4. Walk-Forward Validation Setup
# We start training after we have enough history (e.g., first 2000 fights)
start_index = 2000 
batch_size = 500 # We will predict the next 500 fights, then retrain.

predictions = []
actuals = []
probabilities = []

print(f"Starting Walk-Forward Validation on {len(df)} fights...")

for i in range(start_index, len(df), batch_size):
    # A. Define the Window
    train_data = df.iloc[:i]
    test_data = df.iloc[i : i + batch_size]
    
    if test_data.empty:
        break
        
    X_train = train_data[features]
    y_train = train_data[target]
    
    X_test = test_data[features]
    y_test = test_data[target]
    
    # B. Train the Model (XGBoost)
    # n_estimators=100: Number of trees
    # max_depth=3: Prevents overfitting (UFC data is noisy)
    # learning_rate=0.1: Step size
    model = xgb.XGBClassifier(
        n_estimators=100, 
        max_depth=3, 
        learning_rate=0.1, 
        objective='binary:logistic',
        eval_metric='logloss',
        use_label_encoder=False
    )
    
    model.fit(X_train, y_train)
    
    # C. Predict
    # predict_proba returns [prob_loss, prob_win]. We want [:, 1] for Red Win %
    batch_probs = model.predict_proba(X_test)[:, 1]
    batch_preds = (batch_probs > 0.5).astype(int) # Threshold at 50%
    
    # D. Store Results
    probabilities.extend(batch_probs)
    predictions.extend(batch_preds)
    actuals.extend(y_test.values)
    
    print(f"Batch {i//batch_size}: Trained on {len(train_data)} fights, predicted {len(test_data)}")

# 5. Final Evaluation
overall_acc = accuracy_score(actuals, predictions)
overall_logloss = log_loss(actuals, probabilities)

print("\n--- Final Model Performance ---")
print(f"Accuracy: {overall_acc:.2%}") 
print(f"Log Loss: {overall_logloss:.4f} (Lower is better, <0.69 is edge)")

# 6. Feature Importance (What actually matters?)
import matplotlib.pyplot as plt

xgb.plot_importance(model, importance_type='weight')
plt.title("What Drives the Fight Outcome?")
plt.show()
```
